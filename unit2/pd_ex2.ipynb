```python
import numpy as np
import pandas as pd
```

```python
df=pd.DataFrame({
    "Name":["Aditi","Rohit", "Sneha","Karan",None ,"Alice","Meera","Arjun","Priya",None],
    "Department":["IT","Sales","HR","Sales","Sales", "IT", "Sales", "HR", "IT","HR"],
    "Role": ["Data Analyst", "HR Executive", "Accountant", "Software Engineer",
             "SEO Specialist", "Sales Manager", "ML Engineer",
             "Financial Analyst", "Recruiter", "Content Manager"],
    "Experience": [2, 3, 5, 4, 1, 6, 3, 2, 4, 5],
    "Salary": [450000, 400000, 600000, 750000, 350000,
               800000, 720000, 500000, 420000, 650000],
    "Score": [78, np.nan, 88, 91, np.nan, 89, 93, 81, np.nan, 85]
})
```

```python
df.groupby("Department")["Salary"].mean()
#print(avg)
```

```python
df.groupby("Role")["Score"].mean()
#print(avg_per)
```
