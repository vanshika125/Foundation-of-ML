Handling Missing Values

```python
#updating the dataset with null values
df = pd.DataFrame({
    "Name": ["Aman", "Saurav", "Karan", "Shubham", None, "Sameer"],
    "Age": [22, 23, None, 22, 25, 26],
    "Marks": [85, None, 88, 90, 56, None]
})
```

```python
df.isnull() #checking the missing values
```

```python
df.isnull().sum() #counting the missing values
```

```python
df.dropna() #dropping the missing values
```

```python
print(df)
```

GroupBy in Pandas

```python
data={
    "Name" : ["Aman","Saurav","Karan","Shubham","Abhishek", None, "Sameer"],
    "Age" : [22,23,None, 22, 25, 24, 26],
    "Marks" : [85, None, 88, 90, 56, 72, None],
    "Subject": ["Math", "Physics","Math","Chemistry","Math","Physics","Chemistry"]
}
```

```python
df=pd.DataFrame(data)
```

```python
#basic groupby
df.groupby('Subject')["Marks"].mean() #mean marks per subject . by default they are
```

```python
df.groupby("Subject")["Name"].count() #count students per subject
```

```python
#Multiple aggregations
df.groupby("Subject")["Marks"].agg(["mean","min","max"])
```

```python
df.groupby("Subject")["Marks"].agg(["mean","min","max","count"])
```

```python
#group by multiple columns
df.groupby(["Subject","Age"])["Marks"].mean()
```
