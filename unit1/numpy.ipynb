Copy function - we use this bcz if we directly copy like this b=a then the original array will be changed but by using copy function original remains unaffected

```python
a=np.array([1,2,3])
b=a.copy()#this logic is safe
b[0]=99
print(a)
print(b)
```

```python
a=np.array([1,2,3])
b=a
b[0]=99
print(a)#this logic is not safe
print(b)
```

Where in numpy

```python
arr=np.array([10,-5,20,-3,15])
```

```python
result=np.where(arr<0,0,arr)
print(result)
```

```python
arr=np.array([1,2,3,4,5])
result=np.where(arr%2!=0,-1,arr)
print(result)
```

```python
marks=np.array([10,80,20,45,50])
status=np.where(marks>=50,1,0)
print(status)
```

```python
data=np.array([10,20,30,40,50])
mean_val=np.mean(data)
result=np.where(data>mean_val,mean_val,data)
print(result)
```

Applying (where) condition in 2d array

```python
mat=np.array([[10,60,30],
              [80,20,90]])
```

```python
result=np.where(mat>50,1,0)
print(result)
```
